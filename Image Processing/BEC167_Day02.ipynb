## **Kernal based operations**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.util import random_noise
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/drive/MyDrive/Workshop_Dataset/lena.png")
cv2_imshow(img)
```

```python
kernal = np.array([[1,1,1],[1,1,1],[1,1,1]])
kernal = (1/np.sum(kernal))*kernal  #filter normalization
print(kernal)
dst = cv2.filter2D(img,-1,kernal)
cv2_imshow(dst)
```

```python
kernal = np.ones([15,15])
kernal = (1/np.sum(kernal))*kernal
dst = cv2.filter2D(img,-1,kernal)   #-1 is to flip the kernal
#print(help(cv2.filter2D))
cv2_imshow(dst)
```

```python
#blur filter
kernal = np.array([[1/16,1/8,1/16],[1/8,1/2,1/8],[1/16,1/8,1/16]])
dst = cv2.filter2D(img,-1,kernal)
cv2_imshow(dst)
```

```python
#Sharpening of image
kernal = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
dst = cv2.filter2D(img,-1,kernal)
cv2_imshow(dst)
```

```python
#Convolution
import cv2
import numpy as np
img = cv2.imread("/content/drive/MyDrive/Workshop_Dataset/lenna.png",0)
kernal = np.array([[0,0,0],
                 [0,1,0],
                 [0,0,0]])
height = img.shape[0]
width = img.shape[1]

result = np.zeros((height-2,width-2),np.uint8)
for i in range(0,height-2):
    for j in range(0,width-2):
      temp=0 
      for k in range(0,3):
        for l in range(0,3):
          temp+=kernal[k][l]*img[i+k][j+l]
      result[i][j] = temp

cv2_imshow(result)
print(img.shape,result.shape)
```

```python
#Convolution
import cv2
import numpy as np
img = cv2.imread("/content/drive/MyDrive/Workshop_Dataset/lenna.png",0)
kernal = np.array([[0,0,0],
                 [0,1,0],
                 [0,0,0]])
height = img.shape[0]
width = img.shape[1]
P =0
S =3
rh = ((height-kernal.shape[0]+2*P)//S)+1
rw = ((height-kernal.shape[1]+2*P)//S)+1
result = np.zeros((height-2,width-2),np.uint8)
for i in range(0,rh,S):
    for j in range(0,rw,S):
      temp=0 
      for k in range(0,3):
        for l in range(0,3):
          temp+=kernal[k][l]*img[i+k][j+l]
      result[i//S][j//S] = temp

cv2_imshow(result)
print(img.shape,result.shape)
```

```python
img = cv2.imread("/content/drive/MyDrive/Workshop_Dataset/lenna.png",0)
S = 2
P = 0
K = 2
Height = img.shape[0]
Width = img.shape[1]
print(img.shape)
Rh = ((Height-K+(2*P))//S)+1
Rw = ((Width-K+(2*P))//S)+1
print(Rh,Rw)
result = np.zeros((Rh,Rw),np.uint8)
for i in range(0,Rh):
  for j in range(0,Rw):
    temp=0
    for k in range(0,S):
      for l in range(0,S):
        temp+=img[i*S+k][j*S+l]
    result[i][j] = temp//(S*S)

cv2_imshow(img)
print("...")
cv2_imshow(result)

```
