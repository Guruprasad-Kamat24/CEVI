A multidimensional array of numpy is called as a tensor in Pytorch

All the functions such as creating zeros, matrix multiplication, operations on array can be made on the tensors with some more additional functions and parameteres.

```python
import numpy as np
import torch
```

```python
# initializing a numpy array
a = np.array(1)

# initializing a tensor
b = torch.tensor(1)

print(a)
print(b)
print(b.item())
```

```python
# initializing two arrays
a = np.array(2)
b = np.array(1)
print(a,b)
# addition
print(a+b)

# subtraction
print(b-a)

# multiplication
print(a*b)

# division
print(a/b)
```

```python
# initializing two tensors
a = torch.tensor(2)
b = torch.tensor(1,dtype=float)
print(a,b)
# addition
print(a+b)

# subtraction
print(b-a)

# multiplication
print(a*b)

# division
print(a/b)
```

## Matrix operation

```python
# setting the random seed for numpy and initializing two matrices
np.random.seed(42)
a = np.random.randn(3,3)
b = np.random.randn(3,3)
print('A\n',a,'\n')
print('B\n',b)
```

```python
# matrix addition
print(np.add(a,b), '\n')

# matrix subtraction
print(np.subtract(a,b), '\n')

# matrix multiplication
print(np.dot(a,b), '\n')

# matrix multiplication
print(np.divide(a,b), '\n')

# matrix transpose
print(np.transpose(a))
```

```python
torch.manual_seed(7)
a = torch.randn(3,3)
b = torch.randn(3,3)
print('A\n',a,'\n')
print('B\n',b)
```

```python
# matrix addition
print(torch.add(a,b), '\n')

# matrix subtraction
print(torch.sub(a,b), '\n')

# matrix multiplication
print(torch.mm(a,b), '\n')

# matrix division
print(torch.div(a,b), '\n')

# matrix transpose
print(torch.t(a))
```

### **Concatination of two tensors**

`torch.cat((tensor1, tensor2),dim) `

Note:- 
1. dim = 0 by default vertical concat.
2. dim = 1 for horizontal concat.


these are keeping in mind of 2 dimensional tensor.

```python
# initializing two tensors
a = torch.tensor([[1,2],[3,4]])
b = torch.tensor([[5,6],[7,8]])
print(a, '\n')
print(b,'\n')
# concatenating vertically

print(torch.cat((a,b),dim=1))
print(torch.cat((a,b),dim=0))
```

### AutoGrad

PyTorch uses a technique called automatic differentiation (AutoGrad). It records all the operations that we are performing and replays it backward to compute gradients. This technique helps us to save time on each epoch as we are calculating the gradients on the forward pass itself.

```python
# initializing a tensor
a = torch.ones((2,2), requires_grad=False)
b = a + 5
c = b.mean()
print('A',a)
print('B',b)
print('C',c)
```

```python
# initializing a tensor
torch.manual_seed(8)
a = torch.ones((2,2), requires_grad=True)
b = a + 5
c = b.mean()
print('A',a)
print('B',b)
print('C',c)
```

```python
a = torch.ones((2,2), requires_grad=True)
b = a + 5
c = b.mean()
print('A',a)
print('B',b)
print('C',c)
```
